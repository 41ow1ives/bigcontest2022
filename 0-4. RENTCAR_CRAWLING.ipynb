## 2. 엔젤렌터카 신한카드 페이지 크롤링

```python
path= "C://Users//hjh05//Desktop//mefasol//code//chromedriver.exe"
driver = webdriver.Chrome(path)

url = 'https://jejuangelcar.com/tour.php?p=realcar-list'
driver.get(url=url)
driver.maximize_window()

source = driver.page_source
soup = BeautifulSoup(source, 'html.parser')
cleaning = re.compile('<p.*="">')

driver.implicitly_wait(time_to_wait = 10)
```

```python
#날짜 바꾸기
driver.find_element_by_css_selector('#l-wrap2 > section > form > div > div.searchInline.searchCalendar.datePicker > img').click()
#10월 18일
driver.find_element_by_css_selector('#renewalDatePicker > div > div.ui-datepicker-group.ui-datepicker-group-middle > table > tbody > tr:nth-child(4) > td:nth-child(3) > a').click()
#10월 19일
driver.find_element_by_css_selector('#renewalDatePicker > div > div.ui-datepicker-group.ui-datepicker-group-middle > table > tbody > tr:nth-child(4) > td:nth-child(4) > a').click()

driver.find_element_by_css_selector('#serach_form_submit').click()
driver.find_element_by_css_selector('#l-wrap2 > section > form > div > div.searchInline.searchButton > button').click()
```

```python
# implicit Waits
driver.implicitly_wait(time_to_wait = 5)
```

```python
final_df2 = pd.DataFrame(columns = ['SM_CAT', 'FUEL_TYPE', 'PRICE'])
car_type = 5
```

```python
def click_get_info():
    SM_CAT = driver.find_element_by_css_selector('#info_table > div > div.carItem.w65 > div > p.itemTitle').text
    info = driver.find_element_by_css_selector('#info_table > div > div.carItem.w34 > p.itemTag').text
    FUEL_TYPE = info.split()[0]
    price = driver.find_element_by_css_selector('#info_table > div > div.carPrice > div.priceInline.w48.text-right > p.price > span').text

    return {'SM_CAT' : SM_CAT, 'FUEL_TYPE' : FUEL_TYPE, 'PRICE': price}
```

```python
def scroll():
    SCROLL_PAUSE_TIME = 3
    last_height = driver.execute_script("return document.body.scrollHeight") 

    for j in range(6): #5번 
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") 
        time.sleep(SCROLL_PAUSE_TIME) 
```

```python
scroll()

cars = len([element.text.split('\n') for element in driver.find_elements_by_css_selector('#content > section > ul')][0])//7

for i in range(2,cars+5): #여유를 두고
    try :
        place_click = driver.find_element_by_css_selector(f'#content > section > ul > li:nth-child({i}) > div.buttonList > button.rentButton.direct.bookingButton.rentButtonOpen')
        place_click.click()
        driver.implicitly_wait(time_to_wait = 5)
    
        tmp_info_dict = click_get_info()
        tmp_df = pd.DataFrame(columns = tmp_info_dict.keys(), data = [tmp_info_dict.values()])
        final_df2 = final_df.append(tmp_df)

        driver.back()

        scroll()
    except :
        print("no more info",i)
        break

final_df2
```

```python
driver.quit()
```

```python
finals = pd.concat([final_df,final_df2], ignore_index=True)
finals
```

```python
finals.to_csv('ALLTHAT_제주렌터카.csv', encoding='utf-8-sig', index=False)
```
